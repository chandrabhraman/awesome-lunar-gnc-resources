# Define time period for analysis

```python
'''
  Reference:
    - https://naif.jpl.nasa.gov/pub/naif/toolkit_docs/C/info/mostused.html
'''

step = 300 #(1*365+0)*24

# Reference for mapping day of year to date: https://www.esrl.noaa.gov/gmd/grad/neubrew/Calendar.jsp?view=DOY&year=2009&col=4

# we are going to get positions between these two dates
utc = ['Jun 30, 2009', 'Jul 30, 2009']

# get et values one and two, we could vectorize str2et
etOne = spice.str2et(utc[0])
etTwo = spice.str2et(utc[1])

print("ET One: {}, ET Two: {}".format(etOne, etTwo))

# get times
times = [x*(etTwo-etOne)/step + etOne for x in range(step)]

# check first few times:
#print(times[0:3])
```

# Get position history

```python
positions, lightTimes = spice.spkpos('LRO', times, 'J2000', 'NONE','MOON')
```

# Plot positions

```python
positions = positions.T # positions is shaped (N, 3), let's transpose to (3, N) 
fig = plt.figure(figsize=(9, 9))
ax = fig.add_subplot(111, projection='3d')
ax.plot(positions[0], positions[1], positions[2])
plt.title('SpiceyPy LRO Position (km) from {} to {}'.format(utc[0],utc[1]))
plt.show()
```

# Sun Pointing Analysis

```python
# Sun Pointing Analysis
step = 1500 

# we are going to get positions between these two dates
#utc_spa = ['Jun 30, 2009', 'Jul 20, 2009']
utc_spa = ['Jul 12, 2009', 'Jul 13, 2009']

# get et values one and two, we could vectorize str2et
etOne_spa, etTwo_spa = spice.str2et(utc_spa[0]), spice.str2et(utc_spa[1])

# get times
times_spa = [x*(etTwo_spa-etOne_spa)/step + etOne_spa for x in range(step)]
times_spa_epoch = np.array(times_spa)-times_spa[0]
```

# Function to get Spacecraft clock time

```python
def convtm(et_hist):
    
    SCLKID = -85

    sclk_times = []
    # Convert ET to spacecraft clock time.
    #for et_time in et_hist:
    et_time=et_hist
    calet = spice.etcal( et_time )
    sclkst = spice.sce2c( SCLKID, et_time )
    sclk_times.append(sclkst)

    return sclk_times
```

```python
# Instrument reference: https://naif.jpl.nasa.gov/pub/naif/pds/data/lro-l-spice-6-v1.0/lrosp_1000/data/fk/lro_frames_2010214_v01.tf

#[pos, ltime] = spice.spkpos('SUN', times_spa,'LRO_SA','LT+S','LRO') # Sun position in Solar array frame

e_x,e_y,e_z,ew_x,ew_y,ew_z=[],[],[],[],[],[]

for time_spa in times_spa:

  # Get sun position w.r.t SC in J2000 frame
  [pos, ltime] = spice.spkpos('SUN', time_spa,'J2000','LT+S','LRO') # Sun position in Spacecraft bus frame

  # Non dimensionalize
  # pos= pos / np.linalg.norm(pos, axis=1, keepdims=True)
  pos= pos / np.linalg.norm(pos)
  
  # Attitude and rate function reference: https://naif.jpl.nasa.gov/pub/naif/toolkit_docs/C/cspice/ckgpav_c.html
  sclk_time=convtm(time_spa)
  
  #Cmat, av, ctime =spice.ckgpav(-85000, sclk_time[0], 0, 'J2000')
  Cmat, av, ctime =spice.ckgpav(-85030, sclk_time[0], 0, 'J2000')
  
  # Compute the angular separation
  x_sc = [1,0,0]
  y_sc = [0,-1,0]
  z_sc = [0,0,1]

  # Rotate sun position from J2000 to Spacecraft solar array frame
  Cmat_T=np.transpose(Cmat)
  sunpos_BF=Cmat_T.dot(np.transpose(pos))

  # Angular position error (deg)
  #sep =  [spice.convrt(spice.vsep(sa_normal, pos[x,:]),'RADIANS', 'DEGREES') for x in range(pos.shape[0])]
  #err_x = [spice.convrt(spice.vsep(x_sc, sunpos_BF),'RADIANS', 'DEGREES') for x in range(pos.shape[0])]
  err_y = [spice.convrt(spice.vsep(y_sc, sunpos_BF),'RADIANS', 'DEGREES') for x in range(pos.shape[0])]
  #err_z = [spice.convrt(spice.vsep(z_sc, sunpos_BF),'RADIANS', 'DEGREES') for x in range(pos.shape[0])]

  # Rate error (deg/s)
  err_w_x = spice.convrt(av[0]-0,'RADIANS', 'DEGREES')
  err_w_y = spice.convrt(av[1]-0,'RADIANS', 'DEGREES')
  err_w_z = spice.convrt(av[2]-0,'RADIANS', 'DEGREES')

  # Append list for angle and rate errors
  e_x.append(err_x)
  e_y.append(err_y)
  e_z.append(err_z)
  ew_x.append(err_w_x)
  ew_y.append(err_w_y)
  ew_z.append(err_w_z)

```

## Angle between body frame axes and instantaneous sun direction




```python
plt.plot(times_spa_epoch,np.array(e_y))
plt.xlabel("Time in seconds past {}".format(utc_spa[0]))
plt.ylabel("Separation in deg")
plt.title('Separation between solar panel and sun from {} to {}'.format(utc_spa[0],utc_spa[1]))
plt.show()
```

# Rate errors 

```python
fig, ax = plt.subplots(3, 1)
plt.subplot(3, 1, 1)
plt.plot(times_spa_epoch,np.array(ew_x))
#plt.xlabel("Time in seconds past {}".format(utc_spa[0]))
plt.ylabel("Rate in deg/s")

plt.subplot(3, 1, 2)
plt.plot(times_spa_epoch,np.array(ew_y))
#plt.xlabel("Time in seconds past {}".format(utc_spa[0]))
plt.ylabel("Rate in deg/s")

plt.subplot(3,1, 3)
plt.plot(times_spa_epoch,np.array(ew_z))
plt.xlabel("Time in seconds past {}".format(utc_spa[0]))
plt.ylabel("Rate in deg/s")

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.show()
```
